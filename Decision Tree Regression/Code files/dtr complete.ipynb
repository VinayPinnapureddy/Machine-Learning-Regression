```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
ds=pd.read_csv("temp.csv")
ds.head()
```

```python
ds.drop("LDAPS_PPT2",axis=1,inplace=True)
ds.drop("LDAPS_PPT3",axis=1,inplace=True)
ds.drop("LDAPS_PPT4",axis=1,inplace=True)
ds.head()
```

```python
ds.drop("LDAPS_PPT1",axis=1,inplace=True)
```

```python
ds.head()
```

```python
ds.replace([np.inf, -np.inf], np.nan, inplace=True)
ds.fillna(38, inplace=True)
```

```python
x=ds.iloc[:,2:-2].values
y=ds.iloc[:,-2].values
print(x)
y=y.reshape(len(y),1)
print(y)
```

```python
from sklearn.impute import SimpleImputer
si=SimpleImputer(missing_values=np.nan,strategy="mean")
si.fit(x)
si.fit(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
sc_x=StandardScaler()
x_train=sc_x.fit_transform(x_train)
x_test=sc_x.transform(x_test)
sc_y=StandardScaler()
y_train=sc_y.fit_transform(y_train)
y_test=sc_y.transform(y_test)
```

```python
from sklearn.svm import SVR
svr=SVR()
svr.fit(x_train,y_train)
```

```python
y_pred=sc_y.inverse_transform(svr.predict(x_test))
print(y_pred)
print(sc_y.inverse_transform(y_test))
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_pred,sc_y.inverse_transform(y_test)))
print(r2_score(svr.predict(x_train),y_train))
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr=DecisionTreeRegressor(max_depth=16,min_samples_split=8,min_samples_leaf=5)
dtr.fit(x_train,y_train)
```

```python
y_pred=sc_y.inverse_transform(dtr.predict(x_test))
print(y_pred)
print(sc_y.inverse_transform(y_test))
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_pred,sc_y.inverse_transform(y_test)))
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
y1_pred=sc_y.inverse_transform(lr.predict(x_test))
print(y1_pred)
print(sc_y.inverse_transform(y_test))
```

```python
print(r2_score(y1_pred,sc_y.inverse_transform(y_test)))
print(r2_score(lr.predict(x_train),y_train))
```
