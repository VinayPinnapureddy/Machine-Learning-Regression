```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
sc_y = StandardScaler()
x = sc_x.fit_transform(x)
y = sc_y.fit_transform(y)
print(x)
print(y)
```

```python
from sklearn.svm import SVR
svr=SVR(kernel="rbf",epsilon=0.1,gamma=0.5,C=1000)
svr.fit(x,y)
```

```python
sc_y.inverse_transform(svr.predict(sc_x.transform([[6.5]])))
```

```python
plt.scatter(sc_x.inverse_transform(x),sc_y.inverse_transform(y),color="red")
plt.plot(sc_x.inverse_transform(x),sc_y.inverse_transform(svr.predict(x)),color="blue")
```

```python
from sklearn.metrics import r2_score
print(r2_score(svr.predict(x),y))
```

# SVC using iris data set

```python
import seaborn as sns
df=sns.load_dataset("iris")
df.head()
```

```python
a=df.iloc[:,:-1]
b=df.iloc[:,:-1]
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
print(x)
y=y.reshape(len(y),1)
print(y)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=le.fit_transform(y)
y=y.reshape(len(y),1)
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
print(x_train)
print(y_train)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
from sklearn.metrics import r2_score
print(r2_score(lr.predict(x_test),y_test))
```

```python
from sklearn.svm import SVC
svc=SVC(kernel="rbf",gamma=0.5,C=1)
svc.fit(x_train,y_train)
```

```python
print(r2_score(svc.predict(x_test),y_test))
```
